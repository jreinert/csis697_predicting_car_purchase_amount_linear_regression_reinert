```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import math
```

```python
# load the dataset and prepare for the data for building model
df = pd.read_csv('car_purchase_history.csv', encoding='iso-8859-1')
X = df.iloc[:, 3:4]
Y = df.iloc[:, -1]
X.shape
```

```python
# Create the Training and Test datasets and train the model
x_train, x_test, y_train, y_test = train_test_split (X, Y, test_size = 0.2, random_state = 42)
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
# Predict the results and show the regression line
y_predict = reg.predict(x_test)

plt.scatter(x_test, y_test)
plt.xlabel('Annual Salary', fontsize = 20)
plt.ylabel('Purchase Amount', fontsize = 20)
plt.plot(x_test, y_predict, lw=2, c='orange')
plt.xlim(35000, 100000)
plt.ylim(0, 70000)
plt.show()
```

```python
# Get the R-Squared 
r_squared = reg.score(x_test, y_test)

# Calculate the errors using RMSE 
rmse = math.sqrt(mean_squared_error(y_test, y_predict))

print('slope: ', reg.coef_)
print('intercept: ', reg.intercept_)
print('R-squared: ', r_squared)
print('RMSE: ', rmse)
```

# Predict the purchase amount

```python
salary = 70000
predict_amount = reg.predict([[salary]])
print('Predicted purchase amount: ', predict_amount)
```
